# CLASSIFICATION: PREDICT OUTBREAK PERIOD AND LOW CASES PERIOD

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('bmh')

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import mean_absolute_error
```

```python
# read in pickle
df_iq = pd.read_pickle('./df_iq.pkl')
df_sj = pd.read_pickle('./df_sj.pkl')

# create month variable
df_sj['month'] = df_sj.index.month
df_iq['month'] = df_iq.index.month

# train test split
#sj
sj_train = df_sj.loc[:'2003-04-20']
sj_Xtrain = sj_train.drop('total_cases', axis = 1)
sj_ytrain = sj_train[['total_cases']]

sj_test = df_sj.loc['2003-04-20':]
sj_Xtest = sj_test.drop('total_cases', axis = 1)
sj_ytest = sj_test[['total_cases']]

#iq
iq_train = df_iq.loc[:'2009-06-20']
iq_Xtrain = iq_train.drop('total_cases', axis = 1)
iq_ytrain = iq_train[['total_cases']]

iq_test = df_iq.loc['2009-06-20':]
iq_Xtest = iq_test.drop('total_cases', axis = 1)
iq_ytest = iq_test[['total_cases']]
```

```python
# month by month only
lr = LinearRegression()
month_dums = pd.get_dummies(sj_Xtrain['month'], prefix = 'month')
X = month_dums
y = sj_ytrain.values
lr.fit(X, y)
predicts = lr.predict(X)
plt.plot(predicts)
plt.plot(y)
```

```python
mean_absolute_error(y, predicts)
```

```python
outbreak = []
for i in range(len(y)):
    if y[i] > 1.5 * predicts[i]:
        outbreak.append(1)
    else:
        outbreak.append(0)
lull = []
for i in range(len(y)):
    if y[i] < .75 * predicts[i]:
        lull.append(1)
    else:
        lull.append(0)
over_under = []
for i in range(len(y)):
    if y[i] > 2 * predicts[i]:
        over_under.append('outbreak')
    elif y[i] < .5 * predicts[i]:
        over_under.append('lull')
    else:
        over_under.append('normal')
```

## USE REAL TEST DATA

```python
# Set up test data to fit into model

# read in test data
test = pd.read_csv('./datasets/dengue_features_test.csv', parse_dates=[3])
submission_format = pd.read_csv('./datasets/submission_format.csv')

#split into cities df
test_sj = test[test['city'] == 'sj']
test_iq = test[test['city'] == 'iq']

# FrontFill to Impute into Missing values
test_sj = test_sj.fillna(method = 'ffill')
test_iq = test_iq.fillna(method = 'ffill')

# set index to the dates
test_sj.set_index('week_start_date', drop = True, inplace = True)
test_iq.set_index('week_start_date', drop = True, inplace = True)

# drop non-numerical values
test_sj.drop(['city', 'year', 'weekofyear'], axis = 1, inplace = True)
test_iq.drop(['city', 'year', 'weekofyear'], axis = 1, inplace = True)

# create month variable
test_sj['month'] = test_sj.index.month
test_iq['month'] = test_iq.index.month
```

## SJ

```python
# create monthly curve on all data

lr = LinearRegression()
month_dums = pd.get_dummies(df_sj['month'], prefix = 'month')
X = month_dums
y = df_sj['total_cases'].values
lr.fit(X, y)
month_predicts = lr.predict(X)

# create outbreak strings

outbreak = []
for i in range(len(y)):
    if y[i] > 1.5 * month_predicts[i]:
        outbreak.append(1)
    else:
        outbreak.append(0)
lull = []
for i in range(len(y)):
    if y[i] < .75 * month_predicts[i]:
        lull.append(1)
    else:
        lull.append(0)

# Rand Forest Classifier for outbreak/lull

rfc = RandomForestClassifier()
rfc.fit(df_sj.drop('total_cases', axis = 1), outbreak)
outbreak_preds = rfc.predict(df_sj.drop('total_cases', axis = 1))
outbreak_preds_test = rfc.predict(test_sj)

rfc = RandomForestClassifier()
rfc.fit(df_sj.drop('total_cases', axis = 1), lull)
lull_preds = rfc.predict(df_sj.drop('total_cases', axis = 1))
lull_preds_test = rfc.predict(test_sj)

# model on month dums, outbreak, and lull

lr = LinearRegression()

#set up training data
X = pd.get_dummies(df_sj['month'], prefix = 'month')
X['outbreak'] = outbreak_preds
X['lull'] = lull_preds
y = df_sj['total_cases'].values

# set up test data
X_test = pd.get_dummies(test_sj['month'], prefix = 'month')
X_test['outbreak'] = outbreak_preds_test
X_test['lull'] = lull_preds_test


lr.fit(X, y)

sj_predicts = pd.Series(lr.predict(X_test))
sj_predicts = sj_predicts.apply(lambda x: int(x)).apply(lambda x: 0 if x < 0 else x)
plt.plot(sj_predicts)
```

## IQ

```python
# create monthly curve on all data

lr = LinearRegression()
month_dums = pd.get_dummies(df_iq['month'], prefix = 'month')
X = month_dums
y = df_iq['total_cases'].values
lr.fit(X, y)
month_predicts = lr.predict(X)

# create outbreak strings

outbreak = []
for i in range(len(y)):
    if y[i] > 2 * month_predicts[i]:
        outbreak.append(1)
    else:
        outbreak.append(0)
lull = []
for i in range(len(y)):
    if y[i] < .5 * month_predicts[i]:
        lull.append(1)
    else:
        lull.append(0)

# Rand Forest Classifier for outbreak/lull

rfc = RandomForestClassifier()
rfc.fit(df_iq.drop('total_cases', axis = 1), outbreak)
outbreak_preds = rfc.predict(df_iq.drop('total_cases', axis = 1))
outbreak_preds_test = rfc.predict(test_iq)

rfc = RandomForestClassifier()
rfc.fit(df_iq.drop('total_cases', axis = 1), lull)
lull_preds = rfc.predict(df_iq.drop('total_cases', axis = 1))
lull_preds_test = rfc.predict(test_iq)

# model on month dums, outbreak, and lull

lr = LinearRegression()

#set up training data
X = pd.get_dummies(df_iq['month'], prefix = 'month')
X['outbreak'] = outbreak_preds
X['lull'] = lull_preds
y = df_iq['total_cases'].values

# set up test data
X_test = pd.get_dummies(test_iq['month'], prefix = 'month')
X_test['outbreak'] = outbreak_preds_test
X_test['lull'] = lull_preds_test


lr.fit(X, y)

iq_predicts = pd.Series(lr.predict(X_test))
iq_predicts = iq_predicts.apply(lambda x: int(x)).apply(lambda x: 0 if x < 0 else x)
plt.plot(iq_predicts)
```
