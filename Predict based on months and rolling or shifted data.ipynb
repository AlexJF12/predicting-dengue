# PREDICT RESIDUALS AFTER MONTHLY TREND REMOVED on ROLLING FEATURE DATA

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('bmh')

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import mean_absolute_error
```

```python
# read in pickled features
df_sj = pd.read_pickle('./datasets/all_sj.pkl')
df_iq = pd.read_pickle('./datasets/all_iq.pkl')

total_cases = pd.read_csv('./datasets/dengue_labels_train.csv')
cases_sj = total_cases[total_cases['city'] == 'sj']['total_cases']
cases_iq = total_cases[total_cases['city'] == 'iq']['total_cases']

# create month variable
df_sj['month'] = df_sj.index.month
df_iq['month'] = df_iq.index.month
```

## Train-Validation-Test Split
Because this is time-series data and the features set is dependent on the past, the train test split is very delicate. The validation and test feature set will actually draw from data in the past (rolling mean, std, etc.). Therefore, when creating the model, the entire dataset must be used. But I must be careful not to train the model on data from the 'future'.

```python
# train test split
#sj

# choose split dates
sj_valid_split = '2003-4-20'
sj_test_split = '2008-4-27' # this will split between pre and post submission dates

# split into train, valid, test (no y)
sj_train = df_sj.loc[:sj_valid_split]
sj_Xtrain = sj_train
sj_ytrain = cases_sj[:len(sj_train)]

sj_valid = df_sj.loc[sj_valid_split : sj_test_split]
sj_Xvalid = sj_valid
sj_yvalid = cases_sj[len(sj_train):]

sj_test = df_sj.loc[sj_test_split:]
sj_Xtest = sj_test

print len(sj_train), len(sj_Xtrain), len(sj_ytrain)
print len(sj_valid), len(sj_Xvalid), len(sj_yvalid)
print len(sj_test), len(sj_Xtest)
```

```python
#iq

# choose split dates
iq_valid_split = '2007-7-01'
iq_test_split = '2010-7-01' # this will split between pre and post submission dates

# split
iq_train = df_iq.loc[:iq_valid_split]
iq_Xtrain = iq_train
iq_ytrain = cases_iq[:len(iq_Xtrain)]

iq_valid = df_iq.loc[iq_valid_split : iq_test_split]
iq_Xvalid = iq_valid
iq_yvalid = cases_iq[len(iq_train):]

iq_test = df_iq.loc[iq_test_split:]
iq_Xtest = iq_test

# check the lengths
print len(iq_train), len(iq_Xtrain), len(iq_ytrain)
print len(iq_valid), len(iq_Xvalid), len(iq_yvalid)
print len(iq_test), len(iq_Xtest)
```

## SJ

```python
print len(pd.get_dummies(sj_Xtrain['month'], prefix='month')), len(sj_ytrain.values)
print len(pd.get_dummies(sj_Xvalid['month'], prefix='month')), len(sj_yvalid.values)
```

```python
# sj monthly trend

lr_sj = LinearRegression()
X = pd.get_dummies(sj_Xtrain['month'], prefix='month')
y = sj_ytrain.values

lr_sj.fit(X, y)
sj_residuals_train = y - lr_sj.predict(X)


# on validation data
# note: monthly trend does not need previous weeks data, so this can use the validation set
Xtest = pd.get_dummies(sj_Xvalid['month'], prefix='month')
ytest = sj_yvalid.values
sj_residuals_test = ytest - lr_sj.predict(Xtest)

sj_predicts = lr_sj.predict(Xtest)
plt.plot(sj_predicts)
plt.plot(ytest)
plt.show()
print mean_absolute_error(sj_predicts, ytest)
```

```python
# Find the residuals of the monthly trend

plt.figure(figsize=(16, 4))
plt.plot(sj_residuals_test, label = 'residuals', linewidth = 3)
plt.plot(sj_residuals_test + lr_sj.predict(Xtest), 'g-*', alpha = .4, label = 'true values')
plt.plot(lr_sj.predict(Xtest), alpha = .4, label = 'monthly trend')
plt.title('True Values, Monthly Trend, and Residuals of Dengue Cases in San Juan, Puerto Rico')
plt.legend()
print np.mean(sj_residuals_test)

print len(sj_residuals_test), len(sj_residuals_train)

## PREDICT THE BLUE!!!
```

### ROLLING STATS

Here, I look at which features and at how many weeks to go back can predict the residuals. I plan to slowly add the rolling stats of a feature to a model and iteratively see which new features can be added to the model.

Note, San Juan and Iquitos each require this, as different weather stats can signal different things against the residuals.

```python
sj_train.columns
```

```python
print len(df_sj[5-1:675])
print len(sj_residuals_train[5-1:])
```

```python
# rolling mean of feature to predict residuals

feature = []
rsq_score = []
window = []
mae = []

for i in ['ndvi_ne', u'ndvi_nw', u'ndvi_se', u'ndvi_sw',
       u'precipitation_amt_mm', u'reanalysis_air_temp_k',
       u'reanalysis_avg_temp_k', u'reanalysis_dew_point_temp_k',
       u'reanalysis_max_air_temp_k', u'reanalysis_min_air_temp_k',
       u'reanalysis_precip_amt_kg_per_m2',
       u'reanalysis_relative_humidity_percent',
       u'reanalysis_sat_precip_amt_mm',
       u'reanalysis_specific_humidity_g_per_kg', u'reanalysis_tdtr_k',
       u'station_diur_temp_rng_c',
       u'station_max_temp_c', u'station_min_temp_c', u'station_precip_mm']:
    
    for j in range (5, 60):
        win_len = j
        
        #set up training data
        Xtrain_feat_meanroll = df_sj[i].rolling(window = win_len).mean()[:675]
        Xtrain_feat_stdroll = df_sj[i].rolling(window = win_len).std()[:675]
        Xtrain = pd.concat([Xtrain_feat_meanroll, Xtrain_feat_stdroll], axis = 1)[win_len-1:]
        ytrain = sj_residuals_train[win_len-1:]

#         print len(Xtrain), len(ytrain)

        # set up validation data
        Xvalid_feat_meanroll = df_sj[i].rolling(window = win_len).mean()[675:936]
        Xvalid_feat_stdroll = df_sj[i].rolling(window = win_len).std()[675:936]
        Xvalid = pd.concat([Xvalid_feat_meanroll, Xvalid_feat_stdroll], axis = 1)[win_len-1:]
        yvalid = sj_residuals_test[win_len-1:]

        #check lengths
#         print len(Xvalid), len(yvalid)

        # model it!
        lr_sj_resids = LinearRegression()
        lr_sj_resids.fit(Xtrain, ytrain)
        sj_resid_preds = lr_sj_resids.predict(Xvalid)

        # save info sj residual predictions
        window.append(j)
        feature.append(i)
        rsq_score.append(lr_sj_resids.score(Xvalid, yvalid))
        mae.append(mean_absolute_error(sj_resid_preds, yvalid))
```

```python
# print top features by their r-squared at predicting the residuals alone
pd.DataFrame(zip(feature, rsq_score, window, mae)).sort_values(1, ascending = False).head(20)
```

### SHIFTED

This should be incorporated into the for loop above to see if shifted days back explain the variance in the residuals.

```python
# select which features to use

features = ['station_avg_temp_c']

# Fit shifted stats of features on residuals

lr_sj_resid = LinearRegression()

# for loop to see when is best window (how many weeks back)

scores = []
mae = []
shift_num = []
for i in range(2, 60):
    shift = i
    
    # fit on training
    ss = StandardScaler()
    ss.fit(sj_Xtrain[features])
    Xs = pd.DataFrame(ss.transform(sj_Xtrain[features]))
    Xshift = Xs.shift(shift)
    X = Xshift[shift:]
    y = sj_residuals_train[shift:]

    lr_sj_resid.fit(X, y)
    
    # score on test
    Xstest = pd.DataFrame(ss.transform(sj_Xtest[features]))
    Xshift_test = Xstest.shift(shift)
    X_test = Xshift_test[shift:]
    y_test = sj_residuals_test[shift:]
    
    shift_num.append(i)
    predicts = lr_sj_resid.predict(X_test)
    mae.append(mean_absolute_error(predicts, y_test[1:]))
    scores.append(lr_sj_resid.score(X_test, y_test[1:]))
    
#     plt.plot(predicts)
#     plt.plot(y_test)
#     plt.title(str(i))
#     plt.show()
    
    
plt.plot(shift_num, scores)
plt.title('r-square')
plt.show()
# 7 weeks!

plt.plot(shift_num, mae)
plt.title('mae')
plt.show()
```

### Notes:
* Right now, from my own exploration, there are no features that have a positive r-squared when predicting the residuals.
* a thorough check similar to the rolling stats will be a better method
